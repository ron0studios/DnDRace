### loading and filtering the data

```python
data = [ (d.split(",")[0].strip().lower(), d.split(",")[1].strip().lower()) for d in open("data/online_characters/DnD_characters_May2018.txt").read().splitlines()[1:]]
```

probing data

```python
racecount = {}
for i in data:
    if i[1].lower() in racecount:
        racecount[i[1].lower()] += 1
    else:
        racecount[i[1].lower()] = 1

sorted(racecount.items(),key=lambda x: x[1], reverse=True)[:100]
```

mild racism...

```python
def merge(x): # all "half" races are too human sounding
    if "elf" in x and "half" not in x:
        return "elf"
    if "human" in x and "half" not in x:
        return "human"
    if "halfling" in x: # halfling names are basically human
        return "human"
    if "dwarf" in x and "half" not in x:
        return "dwarf"
    if "orc" in x and "half" not in x:
        return "orc"
    if "dragonborn" in x and "half" not in x:
        return "dragonborn"
    if "gnome" in x and "half" not in x: #gnome names are basically dwarves
        return "dwarf"
    if "goblin" in x and "half" not in x: #goblin names are basically orcs
        return "orc"
    if "ogre" in x and "half" not in x: #ogre names are basically orcs
        return "orc"
    if "dragon" in x:
        return "dragonborn"
    if "kobold" in x:
        return "dragonborn"
    if x == "teifling": #common mispel
        return "tiefling"
    return x


for i in range(len(data)):
    data[i] = (data[i][0], merge(data[i][1]))
```
